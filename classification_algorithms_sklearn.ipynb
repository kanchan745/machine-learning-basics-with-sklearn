#LOGISTIC Regression  
**Logistic regression help in to find the Binary Output . Yes or NO 0/1**

```python
from sklearn.linear_model import LogisticRegression
X = [[1],[2],[3],[4],[5]]#Hour studied Input
y = [0,0,1,1,1]#Pass or Fail Output -> Result
model = LogisticRegression()
model.fit(X,y)
hours = float(input("How many hour you studied-> "))
result = model.predict([[hours]])
if result ==1:
  print (f"Based on hours {hours} , you ll likely to PASS")
else:
  print(f"Based on hours {hours} , you ll likely to FAIL")
```

**Predicting Type of Fruit using KNN based on weight(gm) and its size(cm)**

```python
from sklearn.neighbors import KNeighborsClassifier
X = [
    [180,7],
    [200,7.5],
    [250,8],
    [300,8.5],
    [330,9],
    [360,9.5]
]
# 0 = Apple , 1 = Orange
y = [0,0,0,1,1,1]
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X,y)
weight = float(input("Enter the weight in grams : "))
size = float(input("Enter the size in cm : "))
prediction = model.predict([[weight,size]])[0]# this 0 is for finding  proper no from the list- to find clean no
if prediction == 0:
  print("This is  likely an Apple")
else:
  print(" This is likely an Orange")
```

#Decision Tree
It is a model by which we predict ans by asking a  series of yes/no ques based on data feactures
**syntax**


```python
from sklearn.tree import DecisionTreeClassifier
X = [
    [7,2],#Apple
    [8,3],#Apple
    [9,8],#orange
    [10,9]#orange
]
y = [0,0,1,1] # 0 - Apple , 1 = Orange
model = DecisionTreeClassifier()
model.fit(X,y)
size = float(input("Enter the size in cm : "))
shade = float(input ("Enter the colour shade (1-10): "))
result = model.predict([[size,shade]])[0]
if result == 0:
  print("This is likely an Apple")
else:
  print("This is likely an Orange")
```
